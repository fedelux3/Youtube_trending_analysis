```python
import pandas as pd
import os
```

```python
dir1 = "04.01.20 csv files [fino alle 18.55]\\"
dir2 = "04.01.20 csv files (fede) [dalle 20.00]\\"
dir_r = "20.01.04 csv\\"
```

```python
code = []
f = open('country_codes.txt')
for x in f:
    x = x.strip('\n')
    code.append(x)
```

```python
year = "2020"
month = "01"
day = "04"
```

```python
dfs1 = {} #dizionario df prima cartella
dfs2 = {} #dizionario df seconda cartella
dfs_r = {} #dizionario df uniti
for i in code:
    file = dir1 + str(year) + "." + str(month) + "." + str(day) + "_" + str(i) + "_videos.csv"
    df = pd.read_csv(file)
    dfs1[i] = df #salvo nel dizionario1 i df della prima directory

for i in code:
    file = dir2 + str(year) + "." + str(month) + "." + str(day) + "_" + str(i) + "_videos.csv"
    df = pd.read_csv(file)
    dfs2[i] = df #salvo nel dizionario1 i df della prima directory
    
for i in code:
    dfs_r[i] = pd.concat([dfs1[i],dfs2[i]])
    dfs_r[i].to_csv(dir_r + str(year) + "." + str(month) + "." + str(day) + "_" + str(i) + "_videos.csv", index = False)
```
